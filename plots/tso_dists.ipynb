```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


tso301 = "P29702_301_post.txt"
tso303 = "P29702_303_post.txt"
tso401 = "P29702_401_post.txt"
tsoscm2 = "scmixology2_post.txt"

# Load the data
data301 = np.loadtxt(tso301)
data303 = np.loadtxt(tso303)
data401 = np.loadtxt(tso401)
dataSCM2 = np.loadtxt(tsoscm2)

```

```python
data = {"P29702_301": data301, "P29702_303": data303, "P29702_401": data401, "scmixology2": dataSCM2}

# Make a density plot of the data
sns.kdeplot(data, legend=True).set(title='TSO relative distance from read start/end')
```

```python
tso301bp = "P29702_301_tsoinbp.txt"
tso303bp = "P29702_303_tsoinbp.txt"
tso401bp = "P29702_401_tsoinbp.txt"
tsscmixology2bp = "scmixology2_tsoinbp.txt"

# Load the data
data301bpr = np.loadtxt(tso301bp)
data303bpr = np.loadtxt(tso303bp)
data401bpr = np.loadtxt(tso401bp)
dataSCM2bpr = np.loadtxt(tsscmixology2bp)

data401bp = data401bpr[data401bpr < 200]
data303bp = data303bpr[data303bpr < 200]
data301bp = data301bpr[data301bpr < 200]
dataSCM2bp = dataSCM2bpr[dataSCM2bpr < 200]
data401bp_long = data401bpr[data401bpr >= 200]
data303bp_long = data303bpr[data303bpr >= 200]
data301bp_long = data301bpr[data301bpr >= 200]
dataSCM2bp_long = dataSCM2bpr[dataSCM2bpr >= 200]
```

```python
databp = {"P29702_301": data301bp, "P29702_303": data303bp, "P29702_401": data401bp, "scmixology2": dataSCM2bp}

# Make a density plot of the data
sns.kdeplot(databp, legend=True).set(title='TSO distance (bp) from read start/end', xlabel='bp')

```

```python
# Make piecharts of bp and bp_long data for 301, 303 and 401
labels = '< 200 bp', '>= 200 bp'
sizes301 = [len(data301bp), len(data301bp_long)]
sizes303 = [len(data303bp), len(data303bp_long)]
sizes401 = [len(data401bp), len(data401bp_long)]
sizesscm2 = [len(dataSCM2bp), len(dataSCM2bp_long)]

fig, axs = plt.subplots(2, 2, figsize=(15, 5))

axs[0,0].pie(sizes301, labels=labels, autopct='%1.1f%%')
axs[0,0].set_title('P29702_301')

axs[0,1].pie(sizes303, labels=labels, autopct='%1.1f%%')
axs[0,1].set_title('P29702_303')

axs[1,0].pie(sizes401, labels=labels, autopct='%1.1f%%')
axs[1,0].set_title('P29702_401')

axs[1,1].pie(sizesscm2, labels=labels, autopct='%1.1f%%')
axs[1,1].set_title('scmixology2')
plt.show()
```
